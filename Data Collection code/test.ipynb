```python
all_columns = []
for i in csv:
    df = pd.read_csv(i)
    
    print(f"title : {i}")
    columns = df.columns
    for column in columns:
        if column not in all_columns:
            all_columns.append(column)
        print(f"\t--- {column}")
        
    print(f"len of Shape is : {df.shape}")
    print("\n", "-"*100, "\n")
```

```python
select_columns = ["label", "job_title", "job_description", "job_requirements", "city_name", "job_type", "job_link"]
# select_columns = all_columns
df = pd.DataFrame(columns=select_columns)

for i in csv:
    csv_df = pd.read_csv(i)

    csv_df.rename({"job_name": "job_title", "job_responsibilities": "job_description", "location": "city_name", "city": "city_name", "Employment type": "job_type"}, errors="ignore", axis=1, inplace=True)

    
    csv_df = csv_df[select_columns]

    df = pd.concat([df, csv_df])

    
    print(f"Shape : {df.shape}", df.columns) 
```

```python
def delete_brackets(column):
    data = str(column).split("(")[0].split("/")[0].split("-")[0].split("|")[0]
    return data
```

```python
df["job_title"] = df["job_title"].apply(delete_brackets)
```
