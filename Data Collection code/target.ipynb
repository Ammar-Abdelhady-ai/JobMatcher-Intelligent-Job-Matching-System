```python
import threading

summarizer = ""
def define_summarizer():
    from transformers import pipeline
    global summarizer
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    print("definition Done")
define = threading.Thread(target=define_summarizer)
define.start()
```

```python
import fitz  # PyMuPDF

path = r'E:\nour-1.pdf'
def extract_text_from_pdf(pdf_path):
    text = ''
    pdf_document = fitz.open(pdf_path)
    for page_number in range(pdf_document.page_count):
        page = pdf_document[page_number]
        text += page.get_text()
    return str(text.replace("\xa0", ""))


text_data = extract_text_from_pdf(path)
print(text_data)
```

```python
# path = r"F:\NLP Apple Course\NLP Project\Ammar Abdelhady CV.pdf"
path = r"E:\nour-1.pdf"

cv = str(extract_text_from_pdf(path))
cv = text_data
```

```python
a=len(cv)//3
```

```python
text = [cv[:a], cv[a:2*a], cv[2*a:]]
```

```python
def summarization(text):
    global summ_data
    part = summarizer(text, max_length=150, min_length=30, do_sample=False)
    summ_data.append(part[0]["summary_text"].replace("\xa0", ""))
```

```python
define.join()


summ_data = []

a = threading.Thread(target=summarization, args=(text[0],))
b = threading.Thread(target=summarization, args=(text[1],))
c = threading.Thread(target=summarization, args=(text[-1],))

# Start all threads
a.start()
b.start()
c.start()

# Wait for all threads to finish
a.join()
b.join()
c.join()

# Check if all threads have finished
if not a.is_alive() and not b.is_alive() and not c.is_alive():
    print("ALL Done")
    print(summ_data)
```

```python
data = " .".join(summ_data)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

vectorizer.fit(x)
cv_vect = vectorizer.transform([data])
df_vect = vectorizer.transform(x)
```

```python
outs = []
for i in range(0, len([data])):
    print(f"ID: {i}")
    distances = cosine_similarity(cv_vect[i], df_vect).flatten()
    indices = np.argsort(distances)[::-1]
    for _, j in enumerate(indices[:5]):
        outs.append(y[j])
        print(f"{_} nearest label is {colored(y[j], 'green')}",
             f"similarity: {colored(round(distances[j], 3), 'yellow')}")

print("\n")        
item_counts = Counter(outs)
most_common_item, count = item_counts.most_common(1)[0]
print(f"The most common item is '{colored(most_common_item, 'green')}' with {colored(round(count, 3), 'yellow')} occurrences.")
```
